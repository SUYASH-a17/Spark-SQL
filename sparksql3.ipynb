```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Spark DF 3').getOrCreate()
```

```python
data_path ='/content/drive/MyDrive/Data/'
```

```python
file_path = data_path + "utilization.json"
df_3 = spark.read.json(file_path)
df_3.show()
```

```python
df_3.createOrReplaceTempView("utilization")
```

```python
file_path = data_path + "server_name.csv"
df_s = spark.read.csv(file_path, header=True, inferSchema=True)
df_s.show()
```

```python
df_s.createOrReplaceTempView("server")
```

```python
df_c = spark.sql("SELECT DISTINCT server_id \
                  FROM utilization \
                  ORDER BY server_id")
df_c.show()
```

```python
spark.sql("SELECT min(server_id), max(server_id) \
          FROM utilization").show()
```

```python
df_join = spark.sql("SELECT u.server_id, sn.server_name, u.session_count \
                    FROM utilization u \
                    INNER JOIN server_name sn \
                    ON sn.server_id = u.server_id")
df_join.show()
```
